## Task 1. define and add together two constant values!

- Import TensorFlow under the customary name tf
- Define two constants of arbitrary name
- Give them a value, one of them should be 31 as integer, the other should be 11
- Add them together and store the result of them into the variable `meaning_of_life` 
- (observe, that there are two ways to add things - think "operator overload")
- Let's not leave resources open! (context...)
- Print the `meaning_of_life`! :-P

```python
import tensorflow as tf

tensor1 = tf.constant(31, dtype=tf.int32)
tensor2 = tf.constant(11, dtype=tf.int32)

addition = tensor1 + tensor2

with tf.Session() as sess:
  meaning_of_life = sess.run(addition)

print("The meaning of life is:",meaning_of_life)
```

```python
assert meaning_of_life==42
```

## Task 2.: Use input values!

- Ask for a positive integer from the console (use Python3's default input procedure) and store the value to the `noise_input` variable
- Again define an addition, that adds together the previously added two constants and the recently inputted value
- _Please don't define constants again, reuse them_
- Please observe, that for the "inputs" we have to also define a type
- Execute the operation with the input value and store the result in the variable `more_noise_to_meaning`
- Print the result!

```python
noise_input_variable = int(input("enter a positive integer:"),32)

add1 = tensor1 + tensor2 + noise_input_variable

with tf.Session() as sess:
  more_noise_to_meaning = sess.run(add1)

print(more_noise_to_meaning)
```

```python
assert more_noise_to_meaning>42
```

## Task 3.:  Looking for the variable's value

- Define a _tf_ `Variable` named `noise`
- Don't forget to initialize it's value, to 1
- Define an operation `remove_noise_transform` which from the _previous_ addition operation (that we defined in the prior cell) substracts the value of the `noise` (the tf.Variable)
- With the help of the `tf.assign` function crate an `increase_op`, which increases the value of `noise` with 1
- Use a context and initialize the variables
- In the while cycle assign value to the `result` by running a `tf.session`
    - Retrieve the value of the `remove_noise_transform` and the `increase_op` into the `result` list (in this order!) 
    - Since in our graph the `a_placeholder` is a node, we have to again give `noise_input` to it so that it can operate!


```python
noise = tf.Variable(1, dtype=tf.int32)

remove_noise_transform = add1 - noise

increase_op = tf.assign(noise , noise+1)

init = tf.global_variables_initializer()
result =[1000]

with tf.Session() as sess:
    sess.run(init)
    
    while result[0]>42:
        noise_input = int(input("enter a positive integer:"),32)
        result = sess.run([remove_noise_transform, increase_op])
        print(result)
```

```python
assert result[0]==42
assert result[1]>int(noise_input)

print("Congrats, success!!!!")
```
